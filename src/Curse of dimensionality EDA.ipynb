```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
Computes the distance of the
given row with respect to the remaining
points in the dataset
'''
def compute_min_dist(i, D):
    target_row = D[i,:]
    aux_D = np.delete(D, i, 0)
    def get_dist(row):
        return np.linalg.norm(target_row - row)
    distances = np.apply_along_axis(get_dist, 1, aux_D)
    return distances
```

```python
def nearest_in_kD(ceil, N, k, figname):
    D = np.random.uniform(0, ceil, (N, k))
    min_distances = np.array([np.min(compute_min_dist(i,D)) for i in np.arange(0,D.shape[0])])
    n, bins, patches=plt.hist(min_distances, 20, facecolor='b', alpha=0.75)#, 20, density=True, facecolor='g', alpha=0.75)
    plt.xlabel('Euclidean distances from nearest point')
    plt.ylabel('Number of points')
    plt.title('Distance from nearest point in ' + str(k) + '-dimensional space')
    plt.xlim(0, 100)
    plt.ylim(0, 25)
    plt.grid(True)
    plt.savefig(figname)
    plt.show()
```

```python
nearest_in_kD(100, 100, 2, 'curse_of_dimensionality_dim2')
```

```python
D10 = np.random.uniform(0, 100, (100, 10))
D5 = D10[:,:5]
# D3 = D10[:,:3]
D2 = D10[:,:2]
min_distances10 = np.array([np.min(compute_min_dist(i,D10)) for i in np.arange(0,D10.shape[0])])
min_distances5 = np.array([np.min(compute_min_dist(i,D5)) for i in np.arange(0,D5.shape[0])])
# min_distances3 = np.array([np.min(compute_min_dist(i,D3)) for i in np.arange(0,D3.shape[0])])
min_distances2 = np.array([np.min(compute_min_dist(i,D2)) for i in np.arange(0,D2.shape[0])])
plt.hist(min_distances10, 25, facecolor='r', alpha=1, label='k=10')
plt.hist(min_distances5, 25, facecolor='g', alpha=1, label='k=5')
# plt.hist(min_distances3, 25, facecolor='b', alpha=1, label='k=3')
plt.hist(min_distances2, 25, facecolor='purple', alpha=1, label='k=2')
plt.xlabel('Euclidean distances from nearest point')
plt.ylabel('Number of points')
plt.legend(loc=1)
plt.title('Distance from nearest point in k-dimensional space')
plt.xlim(0, 100)
plt.ylim(0, 25)
plt.grid(True)
plt.savefig("curse_of_dimensionality")
plt.show()
```
